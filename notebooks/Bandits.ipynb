```julia
results=[run_study(MetricStudy(; actions=actions, G=G, n_seeds=500, n_iters=100)) for actions in [UniformActions(),GaussianActions()], G in [NarrowBump(), WideBump(), TwoBumps(), ThreeBumps(), Sinc()]]
```

```julia
plot(plot.(results,:cum_regret)..., layout=(2,5))
```

```julia
plot(plot.(results,:simple_regret)..., layout=(2,5))
```

```julia
b = RandomBandit(; actions=GaussianActions(), outs=Set([:plot]))
rd_result = POMDPs.solve(b, ThreeBumps())
animate(rd_result, "./threebumps_gaussian_rand.gif")
```
